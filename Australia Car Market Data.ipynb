# 1. What is the average price of cars across different brands, and which brands have the highest and lowest average prices?

```python
df=pd.read_csv(r"cars_info.csv",usecols=['Brand','Price'])
```

```python
df.groupby('Brand').mean()
```

```python
average = df.groupby('Brand').mean()
average = average.sort_values(by='Price',ascending=False)
print(average)
```

- The highest average brand is Lamborghini  with 452923.333333
- the lowest average brand is Saab with 8435

# 2.How does the condition of a car (Status) affect its price?

```python
df=pd.read_csv(r"cars_info.csv")
```

```python
sns.barplot(y='Price',x='Status',data=df)
```

- as you can see here demo car has high price
- used cars have low price
- so we can say car status can affect the price

# 3.Is there a correlation between mileage (Kilometers) and the price of a car?

```python
df[['Kilometers','Price']].corr()
```

```python
sns.scatterplot(x='Kilometers', y='Price', data=df)
```

- The scatter plot reveals a negative correlation between the kilometers and the price
- the correlation is negative in between Kilometers and Price

# 4.How does engine capacity (CC) relate to car price and mileage?

```python
sns.scatterplot(x='Kilometers', y='Price',hue='CC', data=df)
```

- if kilometer and CC is low , Price is low 
- if kilometer is low and CC is high the price is High 

```python
correlation_matrix = df[['Kilometers', 'Price', 'CC']].corr()
sns.heatmap(correlation_matrix, annot=True)
```
